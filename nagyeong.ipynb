```python
# ìˆ˜ì¹˜í˜• ë³€ìˆ˜ ìŠ¤ì¼€ì¼ë§ 
from sklearn.preprocessing import StandardScaler

df_bank_ready = df_bank.copy()
df_bank_ready = df_bank_ready.drop('duration', axis=1) # í•„ìš” ì—†ëŠ” í•­ëª© ì œê±°

sc = StandardScaler() # ìŠ¤ì¼€ì¼ë§ í•˜ê¸°
num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous'] # intí˜•ì¸ ë³€ìˆ˜ë“¤ ëª¨ë‘ ìŠ¤ì¼€ì¼ë§
df_bank_ready[num_cols] = sc.fit_transform(df_bank_ready[num_cols]) # ìŠ¤ì¼€ì¼ë§í•œ ë°ì´í„° ë‹¤ì‹œ ëŒ€ìž…. 

df_bank_ready
```

```python
# ë²”ì£¼í˜• ë³€ìˆ˜ ì¸ì½”ë”©

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

df_encoded = pd.DataFrame(encoder.fit_transform(df_bank_ready[cat_cols]).toarray())
df_encoded.columns = encoder.get_feature_names_out(cat_cols)

df_bank_ready = df_bank_ready.drop(cat_cols, axis=1)
df_bank_ready = pd.concat([df_encoded, df_bank_ready], axis=1)

df_bank_ready
```

```python
df_bank_ready['deposit'] = df_bank_ready['deposit'].apply(lambda x: 1 if x == 'yes' else 0)
df_bank_ready
```

```python
from pycaret.classification import *
s = setup(df_bank_ready, target='deposit', session_id = 123)
```

```python
best = compare_models()
```

```python
# pycaret ê²°ê³¼ Gradient Boosting Classifierê°€ ì œì¼ ì¢‹ë‹¤ê³  ë‚˜ì™”ìŒ...
# Gradient Boosting Classifier ì“°ê¸° ìœ„í•´ì„œ ë°ì´í„° ì…‹ ë¶„ë¦¬ 
from sklearn.model_selection import train_test_split
```

```python
feature = df_bank_ready.drop('deposit', axis=1)
target = df_bank_ready['deposit']

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, shuffle=True, random_state=121)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier(verbose=True, n_estimators=1000, random_state=121)
gb_clf.fit(X_train, y_train)
gb_clf.score(X_test, y_test)
```

```python
# íŒŒë¼ë¯¸í„° í•˜ë‚˜ì”© ì¡°ì •í•´ë„ ì ìˆ˜ê°€ ìžê¾¸ ë–¨ì–´ì§€ëŠ”....
# Optuna ì‚¬ìš©....í•´ë³¼ê¹Œ...
import optuna
from sklearn.metrics import accuracy_score

def objective(trial):
    # í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ íƒìƒ‰í•  ë²”ìœ„ ì„¤ì •
    n_estimators = trial.suggest_int('n_estimators', 100, 1000)
    max_depth = trial.suggest_int('max_depth', 5, 50)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)

    # ëª¨ë¸ ìƒì„±
    model = GradientBoostingClassifier(n_estimators=n_estimators,
                                  max_depth=max_depth,
                                  min_samples_split=min_samples_split,
                                  random_state=121, verbose=True)

    # ëª¨ë¸ í›ˆë ¨
    model.fit(X_train, y_train)

    # ì˜ˆì¸¡ ë° ì„±ëŠ¥ í‰ê°€
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    return -accuracy

# 4. Optunaì˜ Study ìƒì„± ë° ìµœì í™” ì‹¤í–‰
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

# 5. ìµœì  í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¶œë ¥
print('Best parameters:', study.best_params)
print('Best Accuracy:', study.best_value)
```
