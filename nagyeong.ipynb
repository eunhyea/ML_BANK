```python
# 수치형 변수 스케일링 
from sklearn.preprocessing import StandardScaler

df_bank_ready = df_bank.copy()
df_bank_ready = df_bank_ready.drop('duration', axis=1) # 필요 없는 항목 제거

sc = StandardScaler() # 스케일링 하기
num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous'] # int형인 변수들 모두 스케일링
df_bank_ready[num_cols] = sc.fit_transform(df_bank_ready[num_cols]) # 스케일링한 데이터 다시 대입. 

df_bank_ready
```

```python
# 범주형 변수 인코딩

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

df_encoded = pd.DataFrame(encoder.fit_transform(df_bank_ready[cat_cols]).toarray())
df_encoded.columns = encoder.get_feature_names_out(cat_cols)

df_bank_ready = df_bank_ready.drop(cat_cols, axis=1)
df_bank_ready = pd.concat([df_encoded, df_bank_ready], axis=1)

df_bank_ready
```

```python
df_bank_ready['deposit'] = df_bank_ready['deposit'].apply(lambda x: 1 if x == 'yes' else 0)
df_bank_ready
```

```python
from pycaret.classification import *
s = setup(df_bank_ready, target='deposit', session_id = 123)
```

```python
best = compare_models()
```

```python
# pycaret 결과 Gradient Boosting Classifier가 제일 좋다고 나왔음...
# Gradient Boosting Classifier 쓰기 위해서 데이터 셋 분리 
from sklearn.model_selection import train_test_split
```

```python
feature = df_bank_ready.drop('deposit', axis=1)
target = df_bank_ready['deposit']

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, shuffle=True, random_state=121)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier(verbose=True, n_estimators=1000, random_state=121)
gb_clf.fit(X_train, y_train)
gb_clf.score(X_test, y_test)
```

```python
# 파라미터 하나씩 조정해도 점수가 자꾸 떨어지는....
# Optuna 사용....해볼까...
import optuna
from sklearn.metrics import accuracy_score

def objective(trial):
    # 하이퍼파라미터를 탐색할 범위 설정
    n_estimators = trial.suggest_int('n_estimators', 100, 1000)
    max_depth = trial.suggest_int('max_depth', 5, 50)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)

    # 모델 생성
    model = GradientBoostingClassifier(n_estimators=n_estimators,
                                  max_depth=max_depth,
                                  min_samples_split=min_samples_split,
                                  random_state=121, verbose=True)

    # 모델 훈련
    model.fit(X_train, y_train)

    # 예측 및 성능 평가
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    return -accuracy

# 4. Optuna의 Study 생성 및 최적화 실행
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

# 5. 최적 하이퍼파라미터 출력
print('Best parameters:', study.best_params)
print('Best Accuracy:', study.best_value)
```
