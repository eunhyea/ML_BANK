# train, test 데이터 나누기

```python
# X_train, X_test, y_train, y_test 나누기
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, shuffle=True, random_state=121)
```

### 수치형 데이터 스케일링

```python
from sklearn.preprocessing import StandardScaler
pd.options.display.float_format = '{:.2e}'.format

sc = StandardScaler()

# 수치형 데이터
num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']

sc.fit(X_train[num_cols]) # fit은 X_train만

# 스케일링 적용
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[num_cols] = sc.fit_transform(X_train[num_cols])  # fit은 X_train만
X_test_scaled[num_cols] = sc.transform(X_test[num_cols])  # transform은 X_train, X_test 모두

```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

# 범주형 데이터 인코딩 (fit은 X_train에 대해서만)
encoded_X_train = encoder.fit_transform(X_train[cat_cols])
encoded_X_test = encoder.transform(X_test[cat_cols])

# 인코딩된 데이터를 데이터프레임으로 변환
encoded_X_train_df = pd.DataFrame(encoded_X_train, columns=encoder.get_feature_names_out(cat_cols), index=X_train.index)
encoded_X_test_df = pd.DataFrame(encoded_X_test, columns=encoder.get_feature_names_out(cat_cols), index=X_test.index)
```

# 수치형 데이터와 범주형 데이터 합치기

```python
# 인코딩된 데이터를 데이터프레임으로 변환하여 열 이름 지정
encoded_X_train_df = pd.DataFrame(encoded_X_train, columns=encoder.get_feature_names_out(cat_cols), index=X_train.index)
encoded_X_test_df = pd.DataFrame(encoded_X_test, columns=encoder.get_feature_names_out(cat_cols), index=X_test.index)

# 인코딩된 범주형 데이터와 스케일링된 수치형 데이터를 합치기
X_train_prepared = pd.concat([X_train_scaled.drop(cat_cols, axis=1), encoded_X_train_df], axis=1)
X_test_prepared = pd.concat([X_test_scaled.drop(cat_cols, axis=1), encoded_X_test_df], axis=1)
```
